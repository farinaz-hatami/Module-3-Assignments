# 2. Weekly graph question

Suppose you wanted to show the nitrate and phosphate level in a water sample.  What are the advantages and disadvantages of showing this as an area plot, as opposed to two separate line graphs?  How would you adjust the graphs shown to improve the presentation?

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series_2 =  100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series_2)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

```python
df = pd.DataFrame({'ser1': time_series, 'ser2': time_series_2}, index = datearray)
df.plot.area()
```

* MY ANSWER:

The graphs show nitrate and phosphate levels generated as two time series across monthly dates and then combined into an area plot.

**Advantages of using an area plot**:

Using an area plot combines both nitrate (ser1) and phosphate (ser2) levels into a single figure, which makes it easier to observe how the two variables change over the same time period. Since both datasets share the same date index, the area plot helps show the overall variation in nutrient levels across months and allows viewers to quickly recognize general trends and relative contribution of each nutrient in the water sample.

**Disadvantages of using an area plot:**

However, in this case the area plot stacks the two series, which can make it difficult to clearly interpret the individual nitrate and phosphate values. The second variable visually depends on the first one, so its true magnitude is harder to estimate compared with the separate line plots created using sns.lineplot(). The filled areas may also make overlapping patterns less clear than simple lines.

**How the graphs could be improved**?

Several adjustments could improve the presentation of these graphs:

1) Add a legend to clearly distinguish nitrate and phosphate levels, especially in the area plot where the variables appear together.

2) Include a descriptive title explaining that the graph shows nutrient levels in a water sample over time.

3) Rotate or format the monthly date labels on the x-axis to improve readability.

3) Use clearer color contrast or transparency in the area plot so both series remain easy to distinguish.

4) Keep consistent labeling so both nutrients can be compared easily on the same scale rather than using separate y-axis descriptions in different plots.

These improvements would make the comparison between nitrate and phosphate levels easier to understand while maintaining the trends shown in the generated data.

# Draw line graphs and/or area graphs when there is date or time data together with numerical data.

```python
import matplotlib.pyplot as plt

# make sure Year is numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# remove missing years
df_time = df.dropna(subset=['Year'])

# average rate per year
yearly_data = df_time.groupby('Year')['Rate_SF_pop'].mean()

# line plot
plt.figure()
yearly_data.plot(marker='o')

plt.xlabel("Year")
plt.ylabel("Average Rate (SF population)")
plt.title("Trend of Health Rate Over Time")

plt.show()
```

```python
plt.figure()

yearly_data.plot.area(alpha=0.5)

plt.xlabel("Year")
plt.ylabel("Average Rate (SF population)")
plt.title("Area Plot of Health Rate Over Time")

plt.show()
```

* My expalnation:

Since the dataset contains time information (Year) together with numerical variables such as Rate_SF_pop, I created line and area graphs to analyze how the data changes over time.

First, I converted the Year column into numeric format and removed rows with missing values to make sure the analysis is accurate. Then, I grouped the data by year and calculated the average Rate_SF_pop for each year. This helps summarize the overall trend instead of looking at individual rows.

The line graph was used to clearly show the trend of health rates over time. From the graph, we can observe whether the rates increase, decrease, or remain stable across different years. Line plots are useful for identifying patterns or gradual changes in the data.

The area graph shows the same information but emphasizes the magnitude of the values over time. It helps visualize how much the rates contribute overall during each year and makes long-term changes easier to see.

Based on these graphs, we can check if there is a trend or drift in the data over time, which is important because large changes across years may affect comparisons between different periods. This analysis helps determine whether the dataset behaves as expected and whether it is suitable for further analysis.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 86-98 of the Storytelling With Data book as best you can.  (The second half of chapter three).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Data Definitions
categories = ['Price', 'Convenience', 'Relationship', 'Service', 'Selection']
our_business = [8.5, 8.0, 4.0, 2.0, 1.8]
ranks = ["1 of 6", "2 of 6", "4 of 6", "6 of 6", "6 of 6"]

# Competitor data (estimated lengths for visual reproduction)
competitors = [
    [7.5, 9.0, 8.8, 4.5, 10.0], # Comp A
    [3.0, 5.0, 4.2, 6.2, 5.5],  # Comp B
    [7.8, 5.5, 3.8, 5.8, 4.0],  # Comp C
    [5.0, 8.2, 5.5, 5.5, 7.5],  # Comp D
    [0.5, 0.8, 6.2, 8.5, 8.0]   # Comp E
]

# Setup Figure and Axes
fig, ax = plt.subplots(figsize=(11, 7))
plt.subplots_adjust(left=0.35, top=0.85) # Room for legend and titles

y_pos = range(len(categories))
blue_hex, gray_hex = '#5D85C3', '#BDBDBD'

# Draw Legend (Top Left)
plt.text(-4.8, -1.0, '■ Our business', color=blue_hex, fontweight='bold', fontsize=11)
comp_names = ['Competitor A', 'Competitor B', 'Competitor C', 'Competitor D', 'Competitor E']
for i, name in enumerate(comp_names):
    plt.text(-4.8, -0.7 + (i * 0.25), f'■ {name}', color='#888', fontsize=10)

#  Draw Main Titles
plt.text(-4.8, -1.5, 'Performance overview', fontsize=20, color='#555')
plt.text(0, -1.0, 'Weighted performance index | ', color='#999', fontsize=11)
plt.text(6.0, -1.0, 'relative rank', color=blue_hex, fontsize=11)

#  Draw Bars and Labels
for i in y_pos:
    # Competitor bars (Gray)
    # They are plotted first so they sit behind the blue bar
    for comp_data in competitors:
        ax.barh(i, comp_data[i], height=0.6, color=gray_hex, alpha=0.4)

    # "Our business" bar (Blue)
    ax.barh(i, our_business[i], height=0.2, color=blue_hex)

    # Category Label (Y-axis)
    ax.text(-0.5, i, categories[i], ha='right', va='center', fontsize=12, color='#444')

    # Rank Label (End of Blue Bar)
    ax.text(our_business[i] + 0.1, i, ranks[i], color=blue_hex, va='center', fontsize=10)

# Final Styling
ax.set_yticks(y_pos)
ax.set_yticklabels([]) # Hide default ticks
ax.invert_yaxis()      # Top-down order
ax.axis('off')         # Clean minimalist look

plt.show()
```

* Explanation:

The code creates a horizontal bar chart to compare the performance of business against five competitors across five categories: Price, Convenience, Relationship, Service, and Selection. Competitor bars are drawn in gray as a background, and business bars are overlaid in blue to highlight its performance. Legends, titles, category labels, and rank annotations are added using plt.text, while default axes and ticks are hidden to create a clean, minimalist look.

The resulting graph provides a clear visual comparison. It shows that business scores highest in Price and Convenience, but lower in Service and Selection, with relative ranks displayed next to each bar. Overall, the chart effectively communicates strengths, weaknesses, and competitive positioning at a glance, making it easy to interpret key insights.
